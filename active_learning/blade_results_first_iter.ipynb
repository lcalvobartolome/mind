## Kappa coefficient

```python
df_en = pd.read_csv("/export/usuarios_ml4ds/lbartolome/Repos/umd/LinQAForge/data/models/filtering/rosie_1_20/df_docs_updated_EN.csv")
df_es = pd.read_csv("/export/usuarios_ml4ds/lbartolome/Repos/umd/LinQAForge/data/models/filtering/rosie_1_20/df_docs_updated_ES.csv")
df_second = pd.read_csv("/export/usuarios_ml4ds/lbartolome/Repos/umd/LinQAForge/data/models/filtering/rosie_1_20/second_eval_annotations.csv")
df_third = pd.read_csv("/export/usuarios_ml4ds/lbartolome/Repos/umd/LinQAForge/data/models/filtering/rosie_1_20/second_eval_annotations_upt.csv")
```

```python
from sklearn.metrics import cohen_kappa_score

merged_df = pd.merge(df_en, df_second[['doc_id', 'second_eval_label']], on='doc_id')
kappa = cohen_kappa_score(merged_df['label'], merged_df['second_eval_label'])

print(f"-- -- KAPPA in English: {kappa}")

differing_rows_en = merged_df[merged_df['label'] != merged_df['second_eval_label']]
differing_rows_en[["doc_id", "text", "label", "second_eval_label"]]
```
