```python
from transformers import ( AutoModelForSequenceClassification, AutoTokenizer)
nli_tokenizer = AutoTokenizer.from_pretrained("potsawee/deberta-v3-large-mnli", use_fast=False)
nli_model = AutoModelForSequenceClassification.from_pretrained("potsawee/deberta-v3-large-mnli")
```

```python
import torch
def _check_entailement(textA, textB, threshold=0.5):
    """
    Compute textual entailment between (textA -> textB) using a 2-class MNLI head.
    
    Returns:
        entail_prob (float), contradict_prob (float), entails_bool (bool)
    """
    try:
        inputs = nli_tokenizer.batch_encode_plus(
            batch_text_or_text_pairs=[(textA, textB)],
            add_special_tokens=True,
            return_tensors="pt",
            truncation=True,
            max_length=512,
        )
        with torch.no_grad():
            logits = nli_model(**inputs).logits  # neutral already removed (2 classes)
            probs = torch.softmax(logits, dim=-1)[0]   # [P(entail), P(contradict)]
            entail_prob = float(probs[0].item())
            contradict_prob = float(probs[1].item())
            entails = entail_prob >= threshold
        return entail_prob, contradict_prob, entails
    except Exception as e:
        print(f"Entailment check failed: {e}")
        return 0.0, 0.0, False
```

```python
df = pd.read_parquet("/export/usuarios_ml4ds/lbartolome/Repos/umd/mind/polylingual_discarded.parquet")
for id, el in df.iterrows():
    print(f"QUESTION: {el['question']}\nSOURCE CHUNK: {el['source_chunk']}\nANSWER: {el['a_s']}\n ENTAILMENT: {_check_entailement(el['a_s'], el['source_chunk'])}\n\n")#\nREASON: {el['reason']}
```
