## Keep documents labelled as "In Domain"

```python
df_all = pd.read_parquet("/export/usuarios_ml4ds/lbartolome/Repos/umd/LinQAForge/data/source/corpus_rosie/passages/translated_stops/df_1.parquet")
```

```python
df_all
```

```python
preds_en = pd.read_csv("/export/usuarios_ml4ds/lbartolome/Repos/umd/LinQAForge/data/models/filtering/rosie_1_20/df_docs_predicted_EN.csv")
preds_es = pd.read_csv("/export/usuarios_ml4ds/lbartolome/Repos/umd/LinQAForge/data/models/filtering/rosie_1_20/df_docs_predicted_ES.csv")
```

```python
keep_en = preds_en[(preds_en.predicted_label == 1.0) | (preds_en.label == 1.0)]
keep_en_ids = keep_en.doc_id.values.tolist()
###
keep_es = preds_es[(preds_es.predicted_label == 1.0) | (preds_es.label == 1.0)]
keep_es_ids = keep_es.doc_id.values.tolist()
```

```python
df_keep = df_all[df_all.doc_id.isin(keep_en_ids + keep_es_ids)]
```

```python
df_keep
```

```python
print((len(preds_en)-len(keep_en))/len(preds_en)*100)
print((len(preds_es)-len(keep_es))/len(preds_es)*100)
```

```python
len(preds_en)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_distances

model = SentenceTransformer("all-mpnet-base-v2", device="cuda")
embeds = model.encode(keep_en["text"].tolist(), show_progress_bar=True)
dists = cosine_distances(embeds)
np.quantile(dists, (0.01, 0.05, 0.1))
```
